# Filtering

## Don't use this. Use loc[]

```python
filt = (df['BetterLife'] == 'Yes')
```

```python
df[filt].head(2)
```

```python
# Or
df[df['BetterLife'] == 'Yes'].head(2)
```

## Better approach using .loc

```python
# Also works!! This is better because we can add second parameter where we can add columns as well

df.loc[filt, ['OpenSource', 'Student']].head(2)
```

```python
filt = (df['Hobbyist'] == 'Yes') & (df['Student'] == 'No')
```

```python
df.loc[filt].head(2)
```

```python
# Negated filter | All data that do not match the filter are returned
df.loc[-filt].head(2)
```

```python
high_salary = (df['ConvertedComp'] > 70000) & (df['Country'] == 'India')
```

```python
df.loc[high_salary, ['LanguageWorkedWith', 'ConvertedComp', 'Country']]
```

### isin()

```python
countries = ['India', 'United States', 'United Kingdom', 'Germany']
filt = df['Country'].isin(countries)
```

```python
df.loc[filt, 'Country']
```

### String Manipulation

```python
# For LanguagesWorkedWith, does the string contains Python. If Nan, replace with False

filt = df['LanguageWorkedWith'].str.contains('Python', na=False)
```

```python
df.loc[filt, 'LanguageWorkedWith']
```
